```python
data = pd.read_json('http://www.siboif.gob.ni/rest/estadisticas?intendencia=Bancos&fecha[min]=2020-03-01&fecha[max]=2020-03-31&tipo_reporte') #, orient='split')
```

```python
data1_activos = data.query("intendencia=='Bancos' &  tipo_institucion=='Bancos'  & institucion in ['AVANZ', 'BAC', 'BANCO ATLÁNTIDA', 'BANCO FICOHSA', 'BANCO LAFISE BANCENTRO', 'BANCORP', 'BANPRO', 'BDF'] & tipo_reporte=='Estado de Situación Financiera (ESF)' & variable_1=='Activo'")
data1_activos
```

```python
fecha_datos = data1_activos.fecha.unique()[0]
fecha_datos
```

```python
data2_activos  = data1_activos[['fecha', 'institucion', 'valor_1']]
#data2_activos=data2_activos.append({'fecha' : fecha_datos , 'institucion' : 'BANCORP', 'valor_1' : 0} , ignore_index=True)
#data2_activos.loc[5]=[ 'Apple', 14, 'Yes' ]
data2_activos = pd.DataFrame(np.insert(data2_activos.values, 5, values=[fecha_datos, 'BANCORP',  ''], axis=0),
                             columns =  ["fecha", "institucion", "valor_1"])

data2_activos
```

```python
data3_activos = data2_activos.pivot(index='fecha', columns='institucion', values='valor_1').reset_index()
data3_activos
```
